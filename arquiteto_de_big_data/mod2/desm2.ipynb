```python
import pandas as pd
from fancyimpute import IterativeImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
path = "C:/Users/jonat/Documents/xpe/arquiteto_de_big_data/mod2/datasets-desm2/dados_jogadores.csv"
```

```python
df = pd.read_csv(path, sep=';', encoding='latin-1')
df
```

```python
# Select numeric columns for imputation
numeric_columns = ['num_vitorias', 'num_derrotas', 'total_partidas']

# Initialize IterativeImputer
mice_imputer = IterativeImputer()

# Fit and transform the numeric columns using IterativeImputer
df[numeric_columns] = mice_imputer.fit_transform(df[numeric_columns])

# The selected numeric columns now contain your data with missing values imputed
```

```python
df
```

### Transformando as colunas num_vitorias num_derrotas total_partidas para inteiro

```python
df['num_vitorias'].fillna(0, inplace=True)
df['num_derrotas'].fillna(0, inplace=True)
df['total_partidas'].fillna(0, inplace=True)

df['num_vitorias'] = df['num_vitorias'].astype(int)
df['num_derrotas'] = df['num_derrotas'].astype(int)
df['total_partidas'] = df['total_partidas'].astype(int)
```

### Transformando a coluna data_nascimento para tipo datetime

```python
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'], format='%d/%m/%Y')
```

```python
df.head(10)
```

## Pergunta 1 
## Qual é o país que está em último lugar no rank da média geral de vitórias? (independente do jogo)



```python
media_vitorias_por_pais = df.groupby('pais')['num_vitorias'].mean()
print(media_vitorias_por_pais)
```

```python
# Encontre o país com a menor média de vitórias
ultimo_pais = media_vitorias_por_pais.idxmin()

print("O país com a menor média de vitórias é:", ultimo_pais)
```

## Pergunta 2
## Qual é a soma de derrotas da Argentina no jogo da Dama?

```python
# Filtra o DataFrame para incluir apenas as linhas com jogo 'Dama' e país 'Argentina'
df_argentina_dama = df[(df['jogo'] == 'Dama') & (df['pais'] == 'Argentina')]

# Calcula a soma de derrotas
soma_derrotas_argentina_dama = df_argentina_dama['num_derrotas'].sum()

print("A soma de derrotas da Argentina no jogo da Dama é:", soma_derrotas_argentina_dama)
```

## Pergunta 3
## Qual é o jogo menos jogado entre o público feminino?


```python
# Filtra o DataFrame para incluir apenas as linhas com gênero 'Feminino'
df_feminino = df[df['genero'] == 'Feminino']

# Agrupa por jogo e conta o número de ocorrências
jogos_menos_jogados = df_feminino['jogo'].value_counts().sort_values()

# Obtém o jogo menos jogado
jogo_menos_jogado = jogos_menos_jogados.index[0]

print("O jogo menos jogado entre o público feminino é:", jogo_menos_jogado)
```

## Qual é o jogo que possui o menor número de partidas?


```python
# Agrupe por jogo e calcule a soma do número total de partidas
jogos_soma_partidas = df.groupby('jogo')['total_partidas'].sum().sort_values()

# Obtém o jogo com o menor número de partidas
jogo_menor_partidas = jogos_soma_partidas.index[0]

print("O jogo com o menor número de partidas é:", jogo_menor_partidas)
```
