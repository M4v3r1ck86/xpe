## Pergunta 1 
## Qual é o país que está em último lugar no rank da média geral de vitórias? (independente do jogo)



```python
# Filtre o DataFrame para encontrar os documentos com num_recomendacoes <= 5
resultados = df_ocorrencia[df_ocorrencia['num_recomendacoes'] <= 5]

# Agrupe os resultados pelo campo 'uf' e conte o número de documentos
agregacao = resultados.groupby('uf').size().reset_index(name='contagem')

# Ordene os resultados de forma descendente
agregacao = agregacao.sort_values(by='contagem', ascending=False)

# Filtrar o DataFrame agregacao para encontrar a contagem para o estado "RN"
contagem_rn = agregacao[agregacao['uf'] == 'RN']['contagem'].values

# Verificar se há contagem para o estado "RN"
if len(contagem_rn) > 0:
    total_contagem_rn = contagem_rn[0]
    print("Total da contagem para o estado 'RN':", total_contagem_rn)
else:
    print("Não há contagem para o estado 'RN'.")

```

```python
# Encontre o índice da linha com a contagem máxima
indice_max_contagem = agregacao['contagem'].idxmax()

# Use o índice para obter o estado com a maior contagem e a contagem correspondente
estado_maior_contagem = agregacao.loc[indice_max_contagem, 'uf']
maior_contagem = agregacao.loc[indice_max_contagem, 'contagem']

# Exiba o estado com a maior contagem e o número de documentos
print("Estado com a maior contagem de documentos:", estado_maior_contagem)
print("Número de documentos:", maior_contagem)

```
