```python
path = "C:/Users/jonat/Documents/xpe/arquiteto_de_big_data/mod2/vacinados_do_virus_chines/country_vaccinations.csv"
```

```python
df = pd.read_csv(path, sep=',', encoding='latin-1')
```

```python

# Formate os números com separadores de milhares
df['total_vaccinations'] = df['total_vaccinations'].apply('{:,.2f}'.format)
df['people_vaccinated'] = df['people_vaccinated'].apply('{:,.2f}'.format)
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].apply('{:,.2f}'.format)
df['daily_vaccinations_raw'] = df['daily_vaccinations_raw'].apply('{:,.2f}'.format)
df['daily_vaccinations'] = df['daily_vaccinations'].apply('{:,.2f}'.format)

# Defina uma função para remover zeros à direita e parte decimal se igual a zero
def remove_trailing_zeros(value):
    parts = value.split('.')
    if len(parts) == 2:
        integer_part, decimal_part = parts
        if decimal_part == '00':
            return integer_part
    return value

# Aplique a função para remover os zeros à direita e parte decimal
df['total_vaccinations'] = df['total_vaccinations'].apply(remove_trailing_zeros)
df['people_vaccinated'] = df['people_vaccinated'].apply(remove_trailing_zeros)
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].apply(remove_trailing_zeros)
df['daily_vaccinations_raw'] = df['daily_vaccinations_raw'].apply(remove_trailing_zeros)
df['daily_vaccinations'] = df['daily_vaccinations'].apply(remove_trailing_zeros)
```

## 1. Quais são os países com maior número de pessoas imunizadas?

```python
# Converter a coluna "date" para o tipo datetime para garantir que as datas sejam tratadas corretamente
df['date'] = pd.to_datetime(df['date'])
```

<!-- 
A função reset_index() é usada para redefinir o índice de um DataFrame do pandas. 
Ela é frequentemente usada após operações de agregação, como a função max(), que nesse caso, resulta em um índice hierárquico. 
A função reset_index() move o índice atual (neste caso, o índice hierárquico) de volta para as colunas e cria um novo índice numérico padrão.

No contexto do código, o reset_index() é usado após a operação max() para transformar o resultado em um DataFrame mais convencional. 
Sem o reset_index(), o resultado da agregação teria um índice hierárquico com os países como primeiro nível de índice e a data como segundo nível, o que pode ser menos conveniente para trabalhar em muitos casos.

Isso torna o DataFrame mais fácil de trabalhar em muitas situações, pois o índice volta a ser uma sequência numérica simples. 
-->

```python
# Encontrar a última data de coleta para cada país
ultimas_datas = df.groupby('country')['date'].max().sort_values(ascending=False).reset_index()
ultimas_datas
```

```python

# Juntar os dados do DataFrame original com as últimas datas
dados_ultima_data = df.merge(ultimas_datas, on=['country', 'date'], how='inner')
dados_ultima_data
```

```python
# pd.options.display.float_format = '{:.2f}'.format

total = df.groupby('country')['total_vaccinations'].max().sort_values(ascending=False).reset_index()
total
```

```python
pais_brasil = total[total['country'] == 'Brazil']
print(pais_brasil)
```

```python

imunizacoes_por_pais = dados_ultima_data[['country', 'total_vaccinations']]
imunizacoes_por_pais
```

```python
# Ordenar os resultados em ordem decrescente com base na soma total de pessoas imunizadas
imunizacoes_por_pais_ordenadas = imunizacoes_por_pais.sort_values('total_vaccinations', ascending=False)
imunizacoes_por_pais_ordenadas.head(10)
```

```python










# Ordenar os resultados em ordem decrescente com base na soma total de pessoas imunizadas
imunizacoes_por_pais_ordenadas = imunizacoes_por_pais.sort_values(by='total_imunizadas', ascending=False)

# Selecione o país com o maior número de pessoas imunizadas
pais_com_mais_imunizadas = imunizacoes_por_pais_ordenadas.head(1)

print(pais_com_mais_imunizadas)

```
